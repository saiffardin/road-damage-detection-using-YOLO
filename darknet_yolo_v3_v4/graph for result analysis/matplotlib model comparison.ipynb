```python
def plot_yolo_2_lines(epochs,
                      plot_title,
                      y_axis_name,
                      line_1_name,
                      line_2_name,
                      line_1_info,
                      line_2_info):

  # plot_title = "Multiclass Comparision"

  plt.figure(figsize=(15, 10))
  plt.title(plot_title, fontsize=19)

  it = epochs//1000 + 1

  # line 1 
  x1 = [(i*1000) for i in range(1,it)]  # i = 1 to 8
  y1 = [line_1_info[i] for i in x1]


  # line 2 
  x2 = [(i*1000) for i in range(1,it)]  # i = 1 to 8
  y2 = [line_2_info[i] for i in x2]



  # Set x axis label.
  plt.xlabel("Epochs", fontsize=19)

  # Set y axis label.
  plt.ylabel(y_axis_name, fontsize=19)


  # Set the x, y axis tick marks text size.
  plt.tick_params(axis='both', labelsize=15)


  # 'bo-' means blue color, round points, solid lines
  plt.plot(x1,y1,'bo-',markersize=10,linewidth=2,label=line_1_name)
  plt.plot(x2,y2,'ro-',markersize=10,linewidth=2,label=line_2_name)


  # zip joins x and y coordinates in pairs - for line 2 only
  for x,y in zip(x2,y2):
    label = "{:.2f}".format(y)

    plt.annotate(label, # this is the text
                (x,y), # this is the point to label
                textcoords="offset pixels", # how to position the text
                xytext=(0,10), # distance from text to points (x,y)
                ha='center', # horizontal alignment can be left, right or center
                fontsize = 'xx-large')



  plt.legend(fontsize="x-large")
  plt.grid(True)

  # Display the plot in the matplotlib's viewer.
  plt.show()
```

```python
plot_yolo_2_lines(epochs = 8000,
                  plot_title  = "Multiclass Comparison (mAP)",
                  y_axis_name = "mAP",
                  line_1_name = "YOLOv3",
                  line_2_name = "YOLOv4",
                  line_1_info = v3_multiclass_epochs_map,
                  line_2_info = v4_multiclass_epochs_map)
```

```python
plot_yolo_2_lines(epochs = 8000,
                  plot_title  = "Multiclass Comparison (F1-Score)",
                  y_axis_name = "F1-Score",
                  line_1_name = "YOLOv3",
                  line_2_name = "YOLOv4",
                  line_1_info = v3_multiclass_epochs_f1_score,
                  line_2_info = v4_multiclass_epochs_f1_score)
```
