```python
import matplotlib.pyplot as plt
```

```python
def plot_yolo(epochs,epochs_info,plot_title):

  first_word = plot_title.split()[0]
  print(first_word)

  plot_mAP = False
  plot_F1 = False

  color = ""

  if (first_word=='mAP'):
    plot_mAP = True
    color = 'bo-'

  else:
    plot_F1 = True
    color = 'ro-'

  print('plot_mAP: ',plot_mAP)
  print('plot_F1: ',plot_F1)



  it = epochs//1000 + 1

  plt.figure(figsize=(15, 10))

  # List to hold x values.
  epochs_x = [(i*1000) for i in range(1,it)]


  # List to hold y values.
  info_y = [epochs_info[i] for i in epochs_info]


  # Set the line chart title and the text font size.
  plt.title(plot_title, fontsize=19)


  # Set x axis label.
  plt.xlabel("Epochs", fontsize=19)

  # Set y axis label.
  plt.ylabel(first_word, fontsize=19)


  # Set the x, y axis tick marks text size.
  plt.tick_params(axis='both', labelsize=15)


  # 'bo-' means blue color, round points, solid lines
  plt.plot(epochs_x,info_y,color,markersize=10,linewidth=2)

  # zip joins x and y coordinates in pairs
  for x,y in zip(epochs_x,info_y):

      label = "{:.2f}".format(y)

      plt.annotate(label, # this is the text
                  (x,y), # this is the point to label
                  textcoords="offset points", # how to position the text
                  xytext=(0,10), # distance from text to points (x,y)
                  ha='center', # horizontal alignment can be left, right or center
                  fontsize = 'xx-large')



  # plt.legend()
  plt.grid(True)

  # Display the plot in the matplotlib's viewer.
  plt.show()
```
